```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('diabetes.csv')

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
#df[8] = df[8].replace(0, -1)

msk = np.random.rand(len(df)) < 0.9
train_and_valid = df[msk]
test = df[~msk]

msk = np.random.rand(len(train_and_valid)) < 0.75
train = train_and_valid[msk]
valid = train_and_valid[~msk]

train_labels = np.array(train[8].as_matrix())
train = train.drop(train.columns[8], axis=1)
train_features = np.array(train.as_matrix())

valid_labels = np.array(valid[8].as_matrix())
valid = valid.drop(valid.columns[8], axis=1)
valid_features = np.array(valid.as_matrix())

test_labels = np.array(test[8].as_matrix())
test = test.drop(test.columns[8], axis=1)
test_features = np.array(test.as_matrix())
```

```python
model = Sequential()
```

```python
model.add(Dense(20, input_dim=test_features.shape[1], kernel_initializer='uniform', activation='tanh'))
model.add(Dropout(0.25))
model.add(Dense(20, activation='tanh'))
model.add(Dropout(0.25))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy', 'mae'])
```

```python
model.fit(train_features, train_labels, epochs=500, verbose=0, batch_size=32, validation_data=(valid_features, valid_labels))
```

```python
score = model.evaluate(test_features, test_labels, batch_size=32)
print(score)
```

```python
pred_y = model.predict_classes(test_features)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_labels, pred_y).ravel()
```
